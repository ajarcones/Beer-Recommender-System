#### 2.- MAPEOS DIRECTOS MANUALES.

Despues del estudio de las diferencias entre 'beer_style' y 'style_name' se identifica la necesidad de registrar a mano los valores de 'style_name' para un subconjunto de 11 ESTILOS DE CERVEZA 'beer_style' DE HASTA 2015 YA QUE NO SE PUEDE CREAR UN ALGORITMO PARA AUTOMATIZAR SU MAPEO por distintas causas:

2.a.- 'style_name' DE 2015 QUE HAN CAMBIADO A OTRO NOMBRE NUEVO DE 'style_name'.

2.a.1.- 'Herbed Spiced Beer' EN 2019 HA CAMBIADO A 'Herb and Spice Beer'

2.a.2.- 'Czech Pilsener' EN 2019 HA CAMBIADO A 'Bohemian Pilsener'

2.a.3.- 'German Pilsener' EN 2019 HA CAMBIADO A 'German Pilsner'

2.a.4.- 'Keller Bier Zwickel Bier' EN 2019 HA CAMBIADO A 'German Kellerbier Zwickelbier'

2.b.- EL ALGORITMO DE MAPEO COMUN posterior, mapping_beer_style_not_found('beer_style', lista_de_'style-name'), NO PUEDE MAPEARLOS BIEN PORQUE existen casos en que existe EMPATE, o sea, el mismo numero de palabras del 'beer_style' de HASTA 2015 que han sido encontradas en mas de un estilo actual 'sytle_name' de BeerAdvocates de 2019.

2.b.1.- 'Pumpkin Ale' DEBERIA SER 'Pumpkin Beer'

2.b.2.- 'Munich Helles' Lager DEBERIA SER 'German Helles'

2.b.3.- 'Maibock Helles Bock' DEBERIA SER 'German Maibock'MANO

2.b.4.- 'American Pale Lager' DEBERIA SER 'American Lager'

2.b.5.- 'Lambic Unblended' DEBERIA SER 'Belgian Lambic'(ver: https://birrapedia.com/enciclopedia-de-la-cerveza/straight--unblended--lambic/s)

2.b.6.- 'Saison Farmhouse Ale' DEBERIA SER 'Belgian Saison'

2.b.7.- 'English Barleywine' DEBERIA SER 'British Barleywine'


```python
bs_to_manually_map_list = ['Herbed Spiced Beer', 'Czech Pilsener', \
                           'German Pilsener', 'Keller Bier Zwickel Bier', \
                           'Pumpkin Ale', 'Munich Helles Lager', 'Maibock Helles Bock', 'American Pale Lager', \
                           'Lambic Unblended', 'Saison Farmhouse Ale', 'English Barleywine']
bs_manually_mapped_list = ['Herb and Spice Beer', 'Bohemian Pilsener', \
                           'German Pilsner', 'German Kellerbier Zwickelbier', \
                           'Pumpkin Beer', 'German Helles', 'German Maibock', 'American Lager', \
                           'Belgian Lambic', 'Belgian Saison', 'British Barleywine']
```

```python
def get_manually_mapped_bs(bs_to_map):
    mapped_bs = ''
    current_bs_index = 0
    for current_bs in bs_to_manually_map_list:
        if current_bs == bs_to_map:
            mapped_bs = bs_manually_mapped_list[current_bs_index]
            break
        else:
            current_bs_index += 1
    return mapped_bs
```

```python
dfreviews[dfreviews['beer_style']\
          .apply(lambda beer_style: beer_style in bs_to_manually_map_list)]['beer_style'].count()
```

COMPROBACION PREVIA DE LOS MAPEOS DIRECTOS MANUALES

```python
for bs in bs_to_manually_map_list:
    print("'{0}' mapped to '{1}'".format(bs, get_manually_mapped_bs(bs)))
```

```python
for bs in bs_to_manually_map_list:
    dfreviews.loc[dfreviews['beer_style'] == bs, 'beer_advocates_style'] = \
        get_manually_mapped_bs(bs)
```

```python
dfreviews[dfreviews['beer_advocates_style'] \
          .apply(lambda beer_advocates_style: beer_advocates_style in bs_manually_mapped_list)] \
         ['beer_advocates_style'].count()
```

```python
# SE HAN MAPEADO 129262 registros CORRECTAMENTE
dfreviews[dfreviews['beer_advocates_style'] \
          .apply(lambda beer_advocates_style: beer_advocates_style in bs_manually_mapped_list)] \
          ['beer_style'] \
          .apply(lambda beer_style: beer_style in bs_to_manually_map_list).count()
```

#### 5.- MAPEO CON UN ALGORITMO COMUN PARA EL RESTO DE LOS 'beer_style'

```python
def words_number_found(input_word, words_to_search_for_list, separator_char):
    words_of_input_word = input_word.split(separator_char)
    #print('words_of_input_word: "{0}"'.format(words_of_input_word))
    counters_dict = {}
    for word_to_search_for in words_to_search_for_list:
        #print('word_to_search_for: "{0}"'.format(word_to_search_for))
        words_of_word_to_search_for = word_to_search_for.split(separator_char)
        found_words_counter = 0
        #print('words_of_word_to_search_for: "{0}"'.format(words_of_word_to_search_for))
        for word in words_of_input_word:
            if word in words_of_word_to_search_for:
                found_words_counter += 1
        #print('found_words_counter: "{0}"'.format(found_words_counter))
        counters_dict[word_to_search_for] = found_words_counter

    return counters_dict
```

```python
def mapping_beer_style_not_found(bs, bslist):
    bsd = words_number_found(bs, bslist, ' ')
    found_bs = ''
    max_bs_counter = 0
    for bs in bsd:
        if bsd[bs] > max_bs_counter:
            max_bs_counter = bsd[bs]
            found_bs = bs
    return found_bs
```

```python
bs_mapped_df = dfreviews[dfreviews['beer_advocates_style'] == '']['beer_style'].to_frame() \
               .sort_values(['beer_style'], ascending=False) \
               .drop_duplicates(subset= ['beer_style'], keep='first')['beer_style'].to_frame()
```
